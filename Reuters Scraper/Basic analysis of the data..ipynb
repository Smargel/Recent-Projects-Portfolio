This is a perfunctory analysis to demonstrate that data had been scrapped from [Reuters](https://www.reuters.com/) had been collected and was analyzable.

```python
import pandas as pd
```

```python
#This file was produced by the scraper found in this same directory
df = pd.read_csv(r"C:\Users\Ebel\Reuters Scraper\reuters.csv")
df.head()
```

```python
#It's possible articles without something listed have an impact on the analysis, we'll fill them in here as another category
df = df.fillna('NoneListed')
```

```python
#Engineering a feature that counts the length of the the title scrapped. 
df['Title_Length'] = df.title.str.len()
```

```python
#Ensuring this feature is correctly engineered
df.head()
```

```python
#This will produce a sorted list of authors by their average title length
df.groupby(df['author']).mean().sort_values(by=['Title_Length'])
```

Apparently Chris Kahn, the polling editor for Reuters, has the lengthiest titles out of the data scraped. Whereas William James, a political correspondent in Westminster, primarily covering "brexit" has the shortest out of non-grouped authors.
